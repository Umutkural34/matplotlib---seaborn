# SUBPLOT NEDİR NE İŞE YARAR ?  

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# 2 satır 2 sütunluk bir düzen oluşturalım ...

fig , axes = plt.subplots(2,2)
```

```python
axes.shape
```

```python
x  = np.arange(0,10)
y = x + 7
```

```python
fig , axes = plt.subplots(1,2,figsize = (13,7)) # 1 satır 2 sütun

axes[0].plot(x,y) # ilk tablo
axes[1].plot(x,y) # ikinci tablo 
;
```

```python
fig , axes = plt.subplots(1 , 3 , figsize = (13,3))

for ax in axes :
    ax.plot(x,y+3)
    
# çok grafik yoksa kullanmak pek mantıklı değildir ...    
```

```python
# 2 satır 2 sütunluk bir düzen oluşturalım ...

fig , axes = plt.subplots(2 , 2 , figsize = (13,7))

axes[0,0].plot(x , y) # 1.satır 1.sütun
axes[0,1].plot(x , x**2) # 1.satır 2.sütun
axes[1,0].plot(x , x**(1/2)) # 2.satır 1.sütun
axes[1,1].plot(x , -x); # 2.satır 2.sütun
```

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1][0].plot(x,x**2)
ax[1,1].plot(x,x**0.5)

plt.tight_layout(); # grafikler arasındaki boşlukları ayarlar
```

```python
# grafiklere başlık ismi verebiliriz ...

fig, axes = plt.subplots(2, 2, figsize=(13, 7), dpi=200)

axes[0, 0].plot(x, y)
axes[0, 0].set_title("X ve Y Değerleri")

axes[0, 1].plot(x, x*3)
axes[1, 0].plot(x, x**3)
axes[1, 1].plot(x, x**2)
axes[1, 1].set_title("X' in Karesi")
axes[1, 1].set_xlabel("X değerleri")

fig.suptitle("GENEL BAŞLIK" , fontsize = 20) ; 
```

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1][0].plot(x,x**2)
ax[1,1].plot(x,x**0.5)

fig.subplots_adjust(
    left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.6,
    hspace=0.6,
    ) # grafikler arasındaki boşluk daha iyi ayarlanabilir bu komutla
```
