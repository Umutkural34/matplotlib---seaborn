```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# BOXPLOT NEDİR ?

```python
df = pd.read_csv("salaries.csv") # veri bilimci maaşları
```

```python
df
```

```python
df.columns = ["calisma_yili","deneyim_seviyesi","calisma_tipi","is_basligi","maas","para_birimi","usd_bazinda_maasi",
             "calisan_rezidansi","uzaktan_calisma_orani","sirket_yeri","sirket_buyuklugu"]
```

```python
df
```

```python
# usd bazinda maaşların uzaktan çalışma oranlarına göre boxplot grafiğini çizdirelim ...

plt.figure(figsize = (13,7) , dpi = 200)
sns.boxplot(data = df , x = "uzaktan_calisma_orani" , y = "usd_bazinda_maasi")
```

```python
# orient komutu ile usd bazinda maaşların uzaktan çalışma oranlarına göre boxplot grafiğini ters şekilde çizdirelim ...

plt.figure(figsize = (13,7) , dpi = 200)
sns.boxplot(data = df  , y = "uzaktan_calisma_orani" , x = "usd_bazinda_maasi"  , orient = "h")
```

```python
# usd bazinda maaşların uzaktan çalışma oranlarına göre boxplot grafiğini çizdirelim ve şirket büyüklüğüne göre gruplandıralım ...


plt.figure(figsize = (13,7) , dpi = 200)
sns.boxplot(data = df , x = "uzaktan_calisma_orani" , y = "usd_bazinda_maasi" , hue = "sirket_buyuklugu")
```

# VİOLİNPLOT NEDİR ?  

```python
df
```

```python
# şirket büyüklüğüne göre dolar bazlı maaşların violinplot ile nasıl dağıldığına bakalım ...

plt.figure(figsize = (13,7) , dpi = 200)
sns.violinplot(data = df , x = "sirket_buyuklugu" , y = "usd_bazinda_maasi")
```

```python
# uzaktan çalımaya göre gruplandırarak şirket büyüklüğüne göre dolar bazlı maaşların violinplot ile nasıl dağıldığına bakalım ...


plt.figure(figsize = (13,7) , dpi = 200)
sns.violinplot(data = df , x = "sirket_buyuklugu" , y = "usd_bazinda_maasi" , hue = "uzaktan_calisma_orani")
```

```python
# quartile komutu ile violinin içindeki çizgiyi daha sade hale getirebiliriz ...

plt.figure(figsize = (13,7) , dpi = 200)
sns.violinplot(data = df , x = "sirket_buyuklugu" , y = "usd_bazinda_maasi" , hue = "uzaktan_calisma_orani" , inner = "quartile")
```
