```python
import numpy as np
import matplotlib.pyplot as plt
```

# LEGEND

```python
x = np.arange(0,10)
```

```python
plt.figure(figsize = (12,6) , dpi = 120)
plt.plot(x , x**2 , label = "X in Karesi") # label kısmına bilgilendirme yazarız
plt.plot(x , x**3 , label = "X in küpü") # label kısmına bilgilendirme yazarız
plt.legend() # legend kısmını göstermek için bu kodu yazarız ...
plt.show();
```

```python
plt.figure(figsize = (12,6) , dpi = 120)
plt.plot(x , x**2 , label = "X in Karesi") # label kısmına bilgilendirme yazarız
plt.plot(x , x**3 , label = "X in küpü") # label kısmına bilgilendirme yazarız
plt.legend(loc = 3) # legend kısmının konumunu göstermek için loc komutunu yazarız ...
plt.show();
```

```python
plt.figure(figsize = (12,6) , dpi = 120)
plt.plot(x , x**2 , label = "X in Karesi") # label kısmına bilgilendirme yazarız
plt.plot(x , x**3 , label = "X in küpü") # label kısmına bilgilendirme yazarız
plt.legend(loc=(1.1,0.75)) # legend kısmının konumunu göstermek için loc komutunu yazarız ...
plt.show();
```

# LİNESTYLE

```python
plt.figure(figsize=(12,6), dpi=120)
plt.plot(x, 2*x + x, linestyle = "--")
plt.plot(x, x**2-x, linestyle = "-.");
```

# LİNEWİDTH VE COLOR

```python
plt.figure(figsize = (13,7) , dpi = 120)
plt.plot(x , x**2 + 4*x , color = "red" , ls = "--" , lw = 3)
plt.plot(x , x**3 - 10*x , color = "yellow" , ls = "-." , lw = 5)
plt.plot(x , 10*x , color = "pink" , ls = ":" , lw = 7)
plt.show();
```

# MARKER

```python
plt.figure(figsize = (13,7))
plt.plot(x , x+10 ,color = "red" , marker = "s" , ms = 10)
plt.plot(x , x+20 , color = "green" , marker = "o" , ms = 20)
plt.plot(x , x+30 , color = "yellow" , marker = "+" , ms = 30)
```

```python
plt.figure(figsize = (13,7))
plt.plot(x , x+10 ,color = "red" , marker = "s" , ms = 10 , markerfacecolor = "yellow" , 
         markeredgewidth = 5 , markeredgecolor = "black" )
plt.plot(x , x+20 , color = "green" , marker = "o" , ms = 20 , markerfacecolor = "green" , 
        markeredgewidth = 7 , markeredgecolor = "gray")
plt.plot(x , x+30 , color = "yellow" , marker = "^" , ms = 15 , markerfacecolor = "red" , 
        markeredgewidth = 3 , markeredgecolor = "blue")
```
