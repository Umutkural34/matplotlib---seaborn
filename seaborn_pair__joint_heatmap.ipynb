# HEATMAP

```python
df = pd.read_csv("Berlin_crimes.csv") # berlindeki suÃ§lar
```

```python
df
```

```python
df.columns = ["yÄ±l" , "semt" , "kod" , "lokasyon" , "soygun" , "sokak_soygunlari" , "yaralanma" , "toplu_saldiri" , "tehdit" , 
             "calinma" , "araba" , "arabadan_calinma" , "bisiklet" ,  "hirsizlik" ,"yangin" , "kundakcilik", "zarar" , "duvar_yazisi" 
             , "ilaclar" , "yerel"]
```

```python
df
```

```python
sayisal_bolumler = df[["soygun" , "sokak_soygunlari" , "yaralanma" , "toplu_saldiri" , "tehdit" , "calinma" , "araba" , 
                       "arabadan_calinma" , "bisiklet" , "hirsizlik" , "yangin" , "kundakcilik" , "zarar" , "duvar_yazisi" ,
                      "ilaclar" , "yerel"]]
```

```python
sayisal_bolumler
```

```python
# korelasyon tablosunu heatmap ile Ã§izelim

korelasyon = sayisal_bolumler.corr()
korelasyon
```

```python
plt.figure(figsize = (20,15) , dpi = 200)
sns.heatmap(korelasyon)
```

```python
# cmap komutu ile renkleri daha aÃ§Ä±k yapabiliriz ...

plt.figure(figsize = (20,15) , dpi = 200)
sns.heatmap(korelasyon , cmap = "magma")
```

```python
# lw komutu ile Ã§izgileri kalÄ±nlaÅŸtÄ±rabiliriz ...

plt.figure(figsize = (20,15) , dpi = 200)
sns.heatmap(korelasyon , cmap = "magma" , lw = 0.5)
```

```python
# annot komutu ile deÄŸerleri kutucuklarÄ±n iÃ§ine yazdÄ±rabiliriz ...

plt.figure(figsize = (20,15) , dpi = 200)
sns.heatmap(korelasyon , cmap = "magma" , lw = 0.5 , annot = True)
```

```python
# fmt komutu ile deÄŸerleri tam sayÄ±ya yuvarlayabiliriz ...

plt.figure(figsize = (20,15) , dpi = 200)
sns.heatmap(korelasyon , cmap = "magma" , lw = 0.5 , annot = True , fmt = ".0f")
```

```python
# center komutu ile kutucuklarÄ±n iÃ§ini koyulaÅŸtÄ±rabiliriz ...

plt.figure(figsize = (20,15) , dpi = 200)
sns.heatmap(korelasyon , cmap = "magma" , lw = 0.5 , annot = True , center = 1)
```
