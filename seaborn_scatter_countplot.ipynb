# SCATTER PLOT NEDİR ? 

```python
df
```

```python
# calisanların deneyim seviyesine göre maaşlarının dağılımına bakalım ...

plt.figure(figsize = (8,6) , dpi = 200)

sns.scatterplot(data = df , x = "deneyim_seviyesi" , y = "usd_bazinda_maasi")
```

```python
# hue komutu ile grafik içinde gruplandırma yapabiliriz ...

plt.figure(figsize = (8,6) , dpi = 200)

sns.scatterplot(data = df , x = "deneyim_seviyesi" , y = "usd_bazinda_maasi" , hue = "sirket_buyuklugu") # sirket buyuklugune göre 
# gruplandırma yapıyoruz
```

```python
# size komutuna herhangi bir kategorik değer atarsak kutucukları boyutlarına göre gruplandırır ...

plt.figure(figsize = (13,7))

sns.scatterplot(data = df , x = "deneyim_seviyesi" , y = "usd_bazinda_maasi" , hue = "sirket_buyuklugu" , size = "calisma_tipi")
```

```python
# palette komutu ile renk , alpha komutu ile grafiğin saydamlığı ayarlayabiliriz ...
# alpha 1 e yaklaştıkca saydamlık artar 0 a yaklaştıkça saydamlık azalır ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.scatterplot(data = df , x = "deneyim_seviyesi" , y = "usd_bazinda_maasi" , hue = "sirket_buyuklugu" , palette = "deep",
               alpha = 0.75);
```

```python
# style komutu ile kutucukların şeklinde gruplandırma yapabiliriz ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.scatterplot(data = df , x = "deneyim_seviyesi" , y = "usd_bazinda_maasi" , hue = "sirket_buyuklugu" , palette = "deep",
               alpha = 0.75 , style = "sirket_buyuklugu" );

# şirketlerin büyüklüğüne göre kutucukların şeklini değiştirdi
```

```python
# deneyim seviyelerini göre countplot grafiği çizelim ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.countplot(data = df , x = "deneyim_seviyesi")
```
