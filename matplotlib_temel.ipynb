# MATPLOTLİB NEDİR ?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = x*5
```

```python
# x ve y nin grafiğini çizdirelim ...

plt.plot(x,y); # en son noktalı virgül koyarak çıkıcak olan metinlerden kurtulabilirsiniz ...
```

```python
x = np.arange(0,10)
y = (x**2) - 4*x
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize = (12,6) , dpi = 250) # grafiğin boyutu (dpi : grafiğin görüntü kalitesi)
plt.plot(x,y)
plt.title("Grafik 1" , size = 20) # grafiğin başlığı ve boyutu 
plt.xlabel("X ekseni" , size = 15) # x ekseninin başlığı ve boyutu
plt.ylabel("Y ekseni" , size = 15) # x ekseninin başlığı ve boyutu
plt.xticks(np.arange(min(x),max(x) + 1 , 1)) # x ekseninde hangi değerleri görmek istersek yazdırabiliriz (her değeri gösterir)
plt.yticks(np.arange(min(y),max(y) , 5)) # y ekseninde hangi değerleri görmek istersek yazdırabiliriz 
#plt.xlim(2,8) # x ekseninde görmek istedğimiz değerlerin aralığı
#plt.ylim(25,35) # y ekseninde görmek istedğimiz değerlerin aralığı
plt.show() # grafiği gösterir
```

# FİGÜR KULLANARAK GRAFİK ÇİZME

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])

axes1.plot(x, y)
axes1.set_xlabel('X Ekseni')
axes1.set_ylabel('Y Ekseni')
axes1.set_title('Büyük')

axes2.plot(x, y)
axes2.set_title('Küçük')
plt.show();
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.65, 0.65, 0.25, 0.25]) 

axes1.plot(x, y)

axes1.set_xlabel('X Ekseni')
axes1.set_ylabel('Y Ekseni')
axes1.set_title('Gerçek Figür')

axes2.plot(x, y)
axes2.set_xlim(3,7)
axes2.set_ylim(0,6)
axes2.set_xlabel('X_1')
axes2.set_ylabel('Y_1')
axes2.set_title('Seçili Alan');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.65, 0.65, 0.25, 0.25]) 
axes3 = fig.add_axes([1.1, 1.1, 0.25, 0.25]) 

axes1.plot(x, y)

axes1.set_xlabel('X Ekseni')
axes1.set_ylabel('Y Ekseni')
axes1.set_title('Gerçek Figür')

axes2.plot(x, y)
axes2.set_xlim(3,7)
axes2.set_ylim(0,6)
axes2.set_xlabel('X_1')
axes2.set_ylabel('Y_1')
axes2.set_title('Seçili Alan')

axes3.plot(x,y);
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,0,0])
axes2 = fig.add_axes([0.5,0.5,0.75,0.75])
axes3 = fig.add_axes([1,1,0.25,0.25])

axes1.plot(x,y)
axes1.set_xlabel('X Ekseni')
axes1.set_ylabel('Y Ekseni')

axes2.plot(x,x*x)
axes2.set_xlabel('X Ekseni')
axes2.set_ylabel('Y Ekseni')

axes3.plot(x,y-5)
axes3.set_xlabel('X Ekseni')
axes3.set_ylabel('Y Ekseni')
```
