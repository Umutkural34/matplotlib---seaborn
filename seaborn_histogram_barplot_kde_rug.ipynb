```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# HİSTOGRAM NEDİR ?

```python
df = pd.read_csv("salaries.csv") # veri bilimci maaşları
```

```python
df
```

```python
df.columns = ["calisma_yili","deneyim_seviyesi","calisma_tipi","is_basligi","maas","para_birimi","usd_bazinda_maasi",
             "calisan_rezidansi","uzaktan_calisma_orani","sirket_yeri","sirket_buyuklugu"]
```

```python
df.info()
```

```python
df
```

```python
# şirketlerin büyüklüklerinin dağılımının histogramını çizelim ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.histplot(data = df , x = "sirket_buyuklugu") 
```

```python
#çalışanların dolar bazında maaşını görelim ...
plt.figure(figsize = (13,7) , dpi = 200)

sns.histplot(data = df , x = "usd_bazinda_maasi")
```

```python
# bins ile kutu sayısını ayarlayabiliriz ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.histplot(data = df , x = "usd_bazinda_maasi" , bins = 10) # 10 kutu
```

```python
# grafiğin dağılımına dair eğri çizmek istersek kde komutunu kullanmalıyız ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.histplot(data = df , x = "usd_bazinda_maasi" , bins = 10 , kde = True)
```

```python
# temel matplotlib komutları ile grafiğin renk ve çizgi ayarlarını belirleyebiliriz ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.set(style = "whitegrid")# arka plan
sns.histplot(data = df , x = "usd_bazinda_maasi" , bins = 10 , kde = True  , lw = 3 , ls = "--",
            color = "gray" , edgecolor = "black")
```

# BAR GRAFİKLERİ VE HİSTOGRAM İLE ARASINDAKİ FARK ?

```python
df
```

```python
# çalışanların deneyim seviyelerine göre ortalama maaşlarının bar grafiğini görelim ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.barplot(data = df , x = "deneyim_seviyesi" , y = "usd_bazinda_maasi")
```

```python
# bar grafiğinde kutunun içindeki çizgiler standart sapmanyı gösterir
```

```python
# şirket büyüklüğüne göre gruplandırarak çalışanların deneyim seviyelerine göre ortalama maaşlarının bar grafiğini görelim ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.barplot(data = df , x = "deneyim_seviyesi" , y = "usd_bazinda_maasi" , 
            hue = "sirket_buyuklugu")
```

# RUGPLOT ve KDEPLOT

```python
df
```

```python
# dolar bazında maaşların rugplot grafiğini görelim ...

plt.figure(figsize = (13,7) , dpi = 200)
sns.rugplot(data = df , x = "usd_bazinda_maasi")
```

```python
# aynısını kdeplot ilede yapabilriz ...

plt.figure(figsize = (13,7) , dpi = 200)
sns.kdeplot(data = df , x = "usd_bazinda_maasi") # boş histogram gibi düşünebiliriz ...
```

```python
# bw_adjust ile çizgilerin dağılımını belirleyebiliriz ...

plt.figure(figsize = (13,7) , dpi = 200)
sns.kdeplot(data=df, x='usd_bazinda_maasi',bw_adjust=0.1)
```

```python
plt.figure(figsize = (13,7) , dpi = 200)
sns.kdeplot(data=df, x='usd_bazinda_maasi',bw_adjust=5)
```

```python
plt.figure(figsize = (13,7) , dpi = 200)
sns.kdeplot(data=df, x='usd_bazinda_maasi',bw_adjust=10)
```

```python
# shade komutu ile gölge kullanabiliriz ...

plt.figure(figsize = (13,7) , dpi = 200)

sns.kdeplot(data=df, x='usd_bazinda_maasi',bw_adjust=10 , shade = True , color = "red")
```

```python
# 2 değişken kullanarak değerlerin nerde yoğunlaştığını görebiliriz ... (değerler sayısal olmalı)

plt.figure(figsize = (13,7) , dpi = 200)

sns.kdeplot(data = df , x = "uzaktan_calisma_orani" , y = "usd_bazinda_maasi")
```
